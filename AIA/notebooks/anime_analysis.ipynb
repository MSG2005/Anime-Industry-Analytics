```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("top_anime_dataset.csv")

df.head()
```

```python
df = df.drop_duplicates()
```

```python
df['episodes'].unique()
```

```python
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
```

```python
df['episodes'] = df['episodes'].fillna(0)
```

```python
df['score'] = pd.to_numeric(df['score'], errors='coerce')
```

```python
df = df.dropna(subset=['score'])
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

```python
df['members'] = pd.to_numeric(df['members'], errors='coerce')
df['favorites'] = pd.to_numeric(df['favorites'], errors='coerce')
```

```python
df.info()
```

```python
df.to_csv("cleaned_anime_dataset.csv", index=False)
```

```python
df['genres'].head(10)
```

```python
df_genres = df.assign(genres=df['genres'].str.split(',')).explode('genres')
```

```python
df_genres['genres'] = df_genres['genres'].str.strip()
```

```python
df_genres.head()
```

```python
genre_scores = df_genres.groupby('genres')['score'].mean().sort_values(ascending=False)
```

```python
genre_scores.head(10)
```

```python
genre_scores = genre_scores.reset_index()
```

```python
genre_scores.head()
```

```python
top_genres = genre_scores.head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_genres, x='score', y='genres')
plt.title("Top Anime Genres by Average Rating")
plt.xlabel("Average Score")
plt.ylabel("Genre")
plt.savefig("top_genres.png")
plt.show()
```

```python
#Genres like Psychological, Mystery, and Action tend to have higher average ratings.
#Streaming platforms could prioritize acquiring or producing anime within these genres to increase viewer engagement and satisfaction.
```

```python
df['episodes'].describe()
```

```python
df_episode = df[df['episodes'] <= 100]
```

```python
df_episode.shape
```

```python
bins = [0, 12, 24, 50, 100]
labels = ['Short (1-12)', 'Seasonal (13-24)', 'Medium (25-50)', 'Long (50+)']

df_episode.loc[:, 'episode_group'] = pd.cut(df_episode['episodes'], bins=bins, labels=labels)
```

```python
df_episode[['episodes','episode_group']].head()
```

```python
episode_scores = df_episode.groupby('episode_group', observed=True)['score'].mean().reset_index()
```

```python
episode_scores
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=episode_scores, x='episode_group', y='score')
plt.title("Average Anime Score by Episode Length")
plt.xlabel("Episode Category")
plt.ylabel("Average Score")
plt.savefig("episode_vs_score.png")  
plt.show()
```

```python
# Anime with 13–24 episodes tend to receive the highest average ratings.
# This suggests that seasonal anime formats balance storytelling quality and viewer engagement effectively.
```

```python
df['studios'].head(10)
```

```python
df_studios = df.assign(studios=df['studios'].str.split(',')).explode('studios')
```

```python
df_studios['studios'] = df_studios['studios'].str.strip()
```

```python
df_studios.head()
```

```python
studio_counts = df_studios['studios'].value_counts().reset_index()
studio_counts.columns = ['studio', 'anime_count']
```

```python
studio_counts.head(10)
```

```python
studio_scores = df_studios.groupby('studios')['score'].mean().reset_index()
```

```python
studio_analysis = pd.merge(studio_scores, studio_counts, left_on='studios', right_on='studio')
```

```python
studio_analysis = studio_analysis[studio_analysis['anime_count'] >= 3]
```

```python
top_studios = studio_analysis.sort_values(by='score', ascending=False).head(10)
top_studios
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_studios, x='score', y='studios')
plt.title("Top Anime Studios by Average Rating")
plt.xlabel("Average Score")
plt.ylabel("Studio")
plt.savefig("top_studios.png")
plt.show()
```

```python
#Certain studios such as MAPPA, Madhouse, and Ufotable consistently produce highly rated anime.
#Streaming platforms could prioritize partnerships with these studios to increase the likelihood of producing successful anime content.
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['score_norm','members_norm','favorites_norm']] = scaler.fit_transform(
    df[['score','members','favorites']]
)
```

```python
df[['score_norm','members_norm','favorites_norm']].head()
```

```python
df['success_score'] = (
    0.5 * df['score_norm'] +
    0.3 * df['members_norm'] +
    0.2 * df['favorites_norm']
)
```

```python
top_anime = df.sort_values(by='success_score', ascending=False).head(10)

top_anime[['title','score','members','favorites','success_score']]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_anime, x='success_score', y='title')

plt.title("Top Successful Anime (Combined Success Score)")
plt.xlabel("Success Score")
plt.ylabel("Anime Title")
plt.savefig("top_success_anime.png")
plt.show()
```

```python
#By combining ratings, popularity, and fan engagement into a unified Success Score, we can identify anime that perform strongly across multiple dimensions.
#These titles represent high-value content that streaming platforms should prioritize for licensing, promotion, or sequel development.
```
